# Your First Python Program

Traditionally, every programming language begins with a simple program called **"Hello, World!"**

This program displays a message on the screen and helps verify that the programming environment is working correctly.

In Python, displaying output is done using the `print()` function.

````python
print("Hello, World!")
````

## Code Explanation

```python
print("Hello, World!")
```

### Explanation

- `print()` is a built-in Python function.
- It displays information on the screen.
- `"Hello, World!"` is a string (text enclosed within quotation marks).

### Output

```
Hello, World!
```

> **Why do we use `print()`?**

The `print()` function is used to display variables, messages, calculations, and program results. It is one of the most commonly used functions while learning Python and debugging programs.

# Python Syntax

Syntax refers to the set of rules that define how Python programs are written.

Python is known for its clean and readable syntax. Unlike many programming languages, Python uses indentation (spaces) to define blocks of code instead of curly braces `{}`.

### Example

```
if condition:
    statement
```

Notice the indentation after the colon (`:`). Proper indentation is mandatory in Python.

````python
print("Learning Python")
print("Learning Machine Learning")
print("Learning Deep Learning")
````

## Explanation

Each `print()` statement is executed one after another, displaying the corresponding message on a new line.

Python executes code from top to bottom unless instructed otherwise using loops, functions, or conditional statements.

# Variables

## What is a Variable?

A variable is a named container used to store data in memory. It allows us to save information that can be accessed and modified throughout a program.

Instead of writing the same value repeatedly, we store it in a variable and reuse it whenever needed.

For example, instead of writing the number `25` multiple times, we can store it in a variable called `age`.

---

## Why Do We Use Variables?

Variables make programs:

- Easier to read
- Easier to maintain
- Easier to update
- More reusable

Variables are one of the fundamental building blocks of every Python program.

> **Real-world Example:**  
> In Machine Learning, variables are commonly used to store values such as learning rate, batch size, number of epochs, image dimensions, and model names.

# Variable Naming Rules

Python follows certain rules when naming variables.

## Rules

- Variable names can contain letters, numbers, and underscores (`_`).
- Variable names cannot start with a number.
- Variable names are case-sensitive.
- Spaces are not allowed.
- Avoid using Python reserved keywords as variable names.

## Good Examples

```python
student_name
learning_rate
batch_size
image_width
```

## Bad Examples

```python
2name
student name
class
for
```

````python
student_name = "Ravi"
learning_rate = 0.001
batch_size = 32

print(student_name)
print(learning_rate)
print(batch_size)
````

# Multiple Variable Assignment

Python allows multiple variables to be assigned in a single line, making code shorter and more readable.

````python
a = b = c = 100

print(a)
print(b)
print(c)
````

# Dynamic Typing

Python is a dynamically typed language, which means you do not need to declare the data type of a variable explicitly.

Python automatically determines the data type based on the assigned value.

````python
value = 100
print(value)

value = "Python"
print(value)

value = 3.14
print(value)
````

Initially, `value` stores an integer.

Later, the same variable stores a string.

Finally, it stores a floating-point number.

Python automatically updates the variable's data type.
