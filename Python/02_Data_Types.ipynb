# What are Data Types?

A **data type** defines the kind of value that a variable can store. Every value in Python belongs to a specific data type.

For example:

- `25` is an integer.
- `3.14` is a floating-point number.
- `"Python"` is a string.
- `True` is a Boolean value.

Python automatically determines the data type based on the assigned value, so there is no need to explicitly declare it.

---

### Why are Data Types Important?

Data types help Python understand how data should be stored, processed, and manipulated.

Different data types support different operations. For example:

- Mathematical operations can be performed on numbers.
- Text operations can be performed on strings.
- Logical operations can be performed on Boolean values.

Choosing the appropriate data type makes programs more efficient, readable, and less prone to errors.

## Real-World Applications

Data types are used extensively in software development and Artificial Intelligence.

Some common examples include:

| Data Type | Example Application |
|------------|--------------------|
| Integer | Number of epochs, batch size |
| Float | Learning rate, model accuracy |
| String | Model name, dataset name |
| Boolean | Training status (`True`/`False`) |
| Complex | Scientific computing and signal processing |

# Integer (`int`)

An **integer** is a whole number without a decimal point. It can be positive, negative, or zero.

Integers are one of the most commonly used data types in Python and are ideal for representing countable values.

---

## Syntax

```python
variable_name = integer_value
```

### Examples

```python
age = 25
temperature = -10
count = 0
```

````python
# Integer examples

age = 25
temperature = -10
count = 0

print(age)
print(temperature)
print(count)
````

## Code Explanation

```python
age = 25
```

Stores the integer value **25**.

```python
temperature = -10
```

Stores a negative integer.

```python
count = 0
```

Stores zero, which is also an integer.

The `print()` function displays the values stored in each variable.

### Output

```
25
-10
0
```

## Where are Integers Used?

Integers are commonly used whenever we need to represent whole numbers.

Examples include:

- Number of students in a class
- Number of products in stock
- Employee ID
- Number of images in a dataset
- Epochs while training a Deep Learning model
- Batch size in Machine Learning

## AI Connection

Integers are widely used in Machine Learning and Deep Learning.

Some examples include:

```python

```

These values are whole numbers and therefore use the **integer (`int`)** data type.

## Did You Know?

Python supports integers of **arbitrary size**, meaning integers are not limited to 32-bit or 64-bit values like in many programming languages. Python automatically allocates more memory when larger integers are needed.

# Floating-Point (`float`)

A **floating-point number**, or **float**, is a number that contains a decimal point. Floats are used to represent values that require fractional or decimal precision.

Unlike integers, floats can store values such as `3.14`, `0.5`, and `-12.75`.

## Syntax

```python
variable_name = float_value
```

### Examples

```python
price = 199.99
accuracy = 98.75
temperature = -5.5
```

````python
# Float examples

price = 199.99
accuracy = 98.75
temperature = -5.5

print(price)
print(accuracy)
print(temperature)
````

## Code Explanation

```python
price = 199.99
```

Stores the decimal value **199.99**.

```python
accuracy = 98.75
```

Stores the floating-point value **98.75**.

```python
temperature = -5.5
```

Stores a negative decimal number.

The `print()` function displays the stored values.

### Output

```
199.99
98.75
-5.5
```

## Where are Floats Used?

Floating-point numbers are commonly used whenever decimal precision is required.

Examples include:

- Product prices
- Scientific measurements
- Temperature readings
- Student GPA or CGPA
- Percentage values
- Distance and speed calculations

## Real-World Example

In Machine Learning and Deep Learning, many important values are stored as floating-point numbers.

Examples include:

```python
learning_rate = 0.001
accuracy = 97.85
loss = 0.245
dropout_rate = 0.5
```

These values require decimal precision and are therefore stored using the **float** data type.

---

## Key Difference: Integer vs Float

| Integer (`int`) | Float (`float`) |
|-----------------|-----------------|
| Stores whole numbers | Stores decimal numbers |
| Example: `10` | Example: `10.5` |
| No decimal point | Contains a decimal point |
| Used for counting | Used for measurements and precision |

# String (`str`)

A **string** is a sequence of characters used to represent text. Strings can contain letters, numbers, symbols, and spaces.

In Python, strings are enclosed within either **single quotes (`' '`)** or **double quotes (`" "`)**.

---

## Syntax

```python
variable_name = "text"
```

### Examples

```python
name = "Alice"
course = "Machine Learning"
language = 'Python'
```

Both single and double quotes are valid for creating strings in Python.

````python
# String examples

name = "Alice"
course = "Machine Learning"
language = 'Python'

print(name)
print(course)
print(language)
````

## Code Explanation

```python
name = "Alice"
```

Stores the text **"Alice"** as a string.

```python
course = "Machine Learning"
```

Stores the text **"Machine Learning"**.

```python
language = 'Python'
```

Stores the text **"Python"** using single quotes.

Python treats both single and double quotes equally when creating strings.

### Output

```
Alice
Machine Learning
Python
```

## Where are Strings Used?

Strings are used whenever textual information needs to be stored.

Examples include:

- Person names
- Email addresses
- File names
- City names
- Programming languages
- Product names
- Website URLs

## Real-World Example

Strings are widely used in Machine Learning and Deep Learning to store descriptive information.

Examples include:

```python

```

These values describe different components of a machine learning model and are therefore stored as strings.

## Single Quotes vs Double Quotes

Python allows both single quotes (`'`) and double quotes (`"`).

```python
language1 = "Python"
language2 = 'Python'
```

Both statements produce the same result.

A common practice is to use **double quotes** for general text and **single quotes** when the string itself contains double quotes, improving readability.

````python
language1 = "Python"
language2 = 'Python'

print(language1)
print(language2)
print(language1 == language2)
````

## Output

```
Python
Python
True
```

Both variables store the same string value. The comparison returns `True` because their contents are identical.

## Important Note

Boolean values are **keywords** in Python.

```python
True

```

Notice that both begin with a capital letter.

The following are **not** Boolean values:

```python
true

"True"

```

- `"True"` and `"False"` are **strings**, not Boolean values.

# Complex (`complex`)

A **complex number** is a number that consists of two parts:

- A **real part**
- An **imaginary part**

In Python, the imaginary part is represented using the letter **`j`**.

The general form of a complex number is:

```python
a + bj
```

where:

- `a` is the real part.
- `b` is the imaginary part.

### Examples

```python
3 + 4j
2 - 5j
-1 + 2j
```

Although complex numbers are one of Python's built-in data types, they are less commonly used in everyday programming compared to integers, floats, strings, and Booleans.

````python
# Complex number examples

complex_num1 = 3 + 4j
complex_num2 = 2 - 5j

print(complex_num1)
print(complex_num2)
````

## Code Explanation

```python
complex_num1 = 3 + 4j
```

Creates a complex number where:

- `3` is the real part.
- `4j` is the imaginary part.

```python
complex_num2 = 2 - 5j
```

Creates another complex number with a negative imaginary component.

The `print()` function displays the complex numbers.

### Output

```
(3+4j)
(2-5j)
```

## Where are Complex Numbers Used?

# Checking Data Types using `type()`

Python provides a built-in function called `type()` that returns the data type of a variable or value.

This function is extremely useful for understanding the kind of data stored in a variable and is commonly used while learning Python, debugging programs, and working with Machine Learning datasets.

---

## Syntax

```python
type(variable_name)
```

or

```python
type(value)
```

### Examples

```python
type(25)
type("Python")
type(3.14)
```

````python
# Checking data types

age = 25
accuracy = 98.75
language = "Python"
is_trained = True
complex_num = 3 + 4j

print(type(age))
print(type(accuracy))
print(type(language))
print(type(is_trained))
print(type(complex_num))
````

## Code Explanation

The `type()` function returns the data type of the value stored in a variable.

```python
type(age)
```

Returns:

```python
<class 'int'>
```

Similarly,

```python
type(accuracy)
```

returns:

```python
<class 'float'>
```

The output of the remaining variables is:

```
<class 'str'>
<class 'bool'>
<class 'complex'>
```

Python represents each data type internally as a class, which is why the output is displayed in the format:

```python
<class 'int'>
```

At this stage, you only need to understand that `type()` helps identify the data type of a variable. The concept of classes will be covered later when we study Object-Oriented Programming.

## Why is `type()` Important?

The `type()` function is useful in many situations, including:

- Understanding the type of data stored in a variable.
- Debugging programs.
- Verifying user input.
- Exploring unfamiliar datasets.
- Writing reusable Python functions.

As programs become larger, checking data types helps prevent unexpected errors.
